Data cleaning and transformations for Dataframe df

```python
#printing the schema of dataframe df
df.info()
```

```python
print("Total number of Null values by each column")
print(df.isnull().sum())
```

```python
df.drop("Generation",axis=1,inplace=True)
```

```python
print("Total number of rows for dataframe df",len(df))
```

```python
df.dropna(inplace=True)  #Dropping Null values of dataframe df to enhance data integrity which can produce accurate results during analysis
print("Total number of rows for dataframe df after dropping null values",len(df))
```

```python
df.isnull().sum()
```

```python
print("Total number of duplicates in dataframe df",df.duplicated().sum())
```

```python
df.drop_duplicates(inplace=True) #Dropping duplicate rows
print("Total number of rows for dataframe df after dropping duplicates values",len(df))
```

```python
df.duplicated().sum()
```

```python
df1.info()  #Schema for Dataframe df1
```

```python
print("Total number of Null values by each column")
print(df1.isnull().sum())
```

```python
print("Total number of rows for dataframe df1",len(df1))
```
